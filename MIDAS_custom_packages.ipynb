```python
%%writefile Midas.py
#cd to PACKAGE_DIR before saving the file

#------------------------------------------------------------------------------
# use on top of the jupyter notebook to import custom functions from Midas module
# import os
# from os.path import expanduser
# HOME_DIR = expanduser("~")
# import sys
# sys.path.append(HOME_DIR+'/packages')
# from Midas import Midas_helper
# helper = Midas_helper()
#------------------------------------------------------------------------------

import pickle
import os
import nltk
import re
import zipfile
import pandas as pd
from nltk.corpus import stopwords
import os
from os.path import expanduser
HOME_DIR = expanduser("~")


class Midas_helper():
    def __init__(self):
        nltk.download("stopwords")
        
    def cd_main_data(self):
        '''Change directory to main data directory
        This is the primary repo of all tables
        '''
        os.chdir('/home/apanoli/main_data')
        
    def load_midas_original_labeled_data(self, num_rows=None):
        '''returns a pandas df from "midas_labeled_data_Q12018.csv"
        n_rows is the number of rows of csv to read
        '''
        self.cd_main_data()
        return pd.read_csv('midas_labeled_data_Q12018.csv', nrows=num_rows)
        
    def cd_class_data(self):
        '''Change directory to class data directory
        This contains class-wise tables for the entire labelled data
        '''
        os.chdir('/home/apanoli/main_data/class_tables')
        
    def cd_train_80_20(self):
        '''Train data split to 80:20
        '''
        os.chdir('/home/apanoli/main_data/test_train_80_20/')
        
    def cd_train_60_20_20(self):
        '''Train data split to 60:20:20
        '''
        os.chdir('/home/apanoli/main_data/train_test_val_60_20_20/')
        
    def save_as_pickle(self, data, filename):
        '''Provide full path to the filename
        data can be anything 
        '''
        with open(filename, 'wb') as f:
            pickle.dump(data, f)

    def load_from_pickle(self, filename):
        '''Provide full path to the filename'''
        with open(filename, 'rb') as f:
            return pickle.load(f)

    def find_common(self, list_of_lists):
        '''Input is a list of lists'''
        return list(set.intersection(*map(set, list_of_lists)))

    def find_unique(self, list_of_lists):
        '''Input is a list of lists'''
        return list(set.difference(*map(set, list_of_list)))

    def preproces_text(self, row):
        '''preprocess the text use df.apply()'''
        stop_words = stopwords.words('english')
        row = re.sub('([^\s\w]|_)+', '', row) # remove all NON alphanumeric EXCEPT white space
        row = re.sub('^(?:(?:[0-9]{2}[:\/,]){2}[0-9]{2,4}|am|pm)$', '', row)
        row = row.replace('\n', ' ') # removes line breaks
        row = row.replace('\t', ' ') # removes tabs - final format is tsv, **important
        # split row text in to list of words, remove STOP words and rejoin
        return ' '.join([w for w in row.split() if not w in stop_words]).lower()
    
    def unzip(self, filename):
        '''unzips the .zip file to current dir
        '''
        with zipfile.ZipFile(filename, 'r') as zip_ref:
            zip_ref.extractall('.')
    
```
